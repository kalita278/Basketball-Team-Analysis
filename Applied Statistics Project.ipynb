* This function summarize the central tendency, dispersion of the data
* Central Tendency is represented by min, 25%, 50%, 75% and max
* 25% is also known as First Quartile (Q1), 50% as Second Quartile or Median (Q2) and 75% as Third Quartile (Q3)
* To understand te data better, let's derive the following feature
   * CancelGames: Number of Games played but no result have been declared. These are the games could have cancelled or withdrawn without declaring any result such as won/lost/drawn
   * WinPer: Percentages of games owned by the team
   * LostPer: Percentages of games lost by the team
   * WinRatio: Ratio of games won to that of lost
   * BasketRatio: Ratio of baskets scored to that of given

```python
df['CancelGames']=df['PlayedGames']-(df['WonGames']+df['DrawnGames']+df['LostGames']) #Adding Canceled game feature
df['WinPer'] = df['WonGames']/df['PlayedGames'] #Adding winning percentage feature
df['LostPer'] = df['LostGames']/df['PlayedGames'] #Adding Losing percentage feature
df['WinRatio'] = df['WonGames']/df['LostGames'] #Adding winning ratio feature
df['BasketRatio'] = df['BasketScored']/df['BasketGiven'] #Adding Basket ratio feature
```

```python
#Tournaments played by Teams
plt.figure(figsize=(20,8))
plt.bar(df['Team'],df['Tournament'])
plt.xticks(rotation=90)
plt.xlabel("Teams")
plt.ylabel("No of Tournaments")
```

* Team 1 to Team 8 had played more than 60 tournaments
* The distribution shows that the data is right skewed

```python
#No of mathes won by Teams
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
wg = df[['Team','WonGames']].sort_values('WonGames',ascending=False)
wg1 = wg[0:15]
plt.pie(wg1['WonGames'],labels=wg1['Team'])
plt.title('Matches Won by top 15 Teams')
#No of mathes Lost by Teams
plt.subplot(1,2,2)
wg2 = df[['Team','LostGames']].sort_values('LostGames',ascending=False)
wg3=wg2[0:25]
plt.pie(wg3['LostGames'],labels=wg3['Team'])
plt.title('Matches Lost by top 25 Teams')
```

* From the pie chart, we can see that Team 1 and Team 2 have won the mmost number of games they played and Team 5 and Team 7 have lost most number of games compared to other teams


```python
#Scores of Teams
plt.figure(figsize=(20,8))
plt.bar(df['Team'],df['Score'])
plt.xticks(rotation=90)
plt.xlabel("Teams")
plt.ylabel("Score")
```

* From the bar chart, we can observe that Team 1 and Team 2 have highest scores compared to other teams

```python
#Most number of champions/runner ups
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
champ = df[['Team','TournamentChampion']].sort_values('TournamentChampion',ascending=False)
champ1 = champ[0:10]
plt.bar(champ1['Team'],champ1['TournamentChampion'])
plt.title('Top 10 champion team')
plt.subplot(1,2,2)
runner = df[['Team','Runner-up']].sort_values('Runner-up',ascending=False)
runner1= runner[0:10]
plt.bar(runner1['Team'],runner1['Runner-up'])
plt.title('Top 10 runner up team')
```

* From the First chart, we can see that Team 1 and Team 2 is most number of champion team
* From the Second chart, we can see that Team 2 and Team 1 is the most number of Runner-up Team

```python
#Top 10 team with highest win percentage/Win ratio
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
wp = df[['Team','WinPer']].sort_values('WinPer',ascending=False)
wp1 = wp[0:10]
plt.bar(wp1['Team'],wp1['WinPer']*100)
plt.title('highest win percentage of top 10 teams')
#No of mathes Lost by Teams
plt.subplot(1,2,2)
wr = df[['Team','WinRatio']].sort_values('WinRatio',ascending=False)
wr1=wr[0:10]
plt.bar(wr1['Team'],wr1['WinRatio'])
plt.title('highest win ratio of top 10 teams')
```

* From the First chart, we can see that Team 1 and Team 2 have highest win percentage than other teams
* From the Second chart, we can also see that Team 1 and Team 2 have highest win ratio than other teams
